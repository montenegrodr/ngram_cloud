```python
with open('../data.csv') as h:
    data = [_ for _ in csv.DictReader(h)]
```

```python
print(data[0].keys())
print(data[0]['title'])
print(data[0]['complaint_body'])
```

```python
from nltk import ngrams
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
```

```python
from collections import defaultdict

stpwords = stopwords.words('portuguese')

def sanitize(text):
    return [
        w.strip().lower() for w in text 
        if w.strip().lower() not in stpwords
    ]
    

n = 2
bigrams = defaultdict(int)

for i in range(100000):
    for ngram in ngrams(sanitize(data[i]['final_answer'].split()), n):
        bigrams[ngram] += 1

        
for bigram in sorted(bigrams.items(), key=lambda x: x[1], reverse=True)[:100]:
    print(bigram)
```

```python
# a = sorted(bigrams.items(), key=lambda x: x[1], reverse=True)[:1000]
# for bigram in sorted(a , key=lambda x: x[0], reverse=True)[:1000]:
#     print(bigram)

sorted_bigrams = sorted(bigrams.items(), key=lambda x: x[1], reverse=True)
ranking = defaultdict(int)
for item in sorted_bigrams:
    ranking[item[0][0]] += 1
    ranking[item[0][1]] += 1

 
    
```

```python
ranking = sorted(ranking.items(), key=lambda x: x[1], reverse=True)    
print(ranking)
```
